# Vacinação COVID-19 - 1

A vacinação contra o COVID-19 é a campanha mais aguardada para 2021. Existe muita pressão em cima dos governantes para que isso ocorra da forma mais ágil possível. Todos os estados brasileiros já estão na caminhada e os dados referentes a isso estão disponíveis no OpenDataSUS. Sua meta é baixar o CSV com informações sobre a vacinação no estado de São Paulo e, usando os conceitos aprendidos sobre leitura de arquivos CSV e lógica de programação em geral (estruturas condicionais, estruturas de repetição, manipulação de strings, entre outros), fazer uma mini análise exploratória em cima dessa base, respondendo a seguinte pergunta:

As vacinas aplicadas foram apenas a Sinovac ou existem outras? Caso existam outras, quais são?

A base referente aos estado de São Paulo está disponível no link(https://s3-sa-east-1.amazonaws.com/ckan.saude.gov.br/PNI/vacina/uf/2021-03-12/uf%3DSP/part-00000-f3fe3224-d992-40c3-a6e1-2db8d0be55a7.c000.csv). Tal base foi encontrada por meios das páginas 1 e 2.

```python
#Para ler o arquivo sem necessidade de fechá-lo e fazer a leitura da primeira linha. Utiliza-se o enumerate para saber quais
#são as colunas que estão no cabeçalho
with open('vacina.csv', encoding = 'iso-8859-1') as arq:

    linha = arq.readline()    
    linha = linha.split(',')
    
    for i, l in enumerate(linha):
        print(i, l)  
```

```python
with open('vacina.csv', encoding = 'iso-8859-1') as arq:
    for i in range(5):
        #verifica-se os dados na posição 5
        linha = arq.readline()    
        linha = linha.split(',')
    
        for i, l in enumerate(linha):
            print(i, l)  
```

```python
#Criar lista vazia para armazenar os tipos diferentes de vacinas e faz leitura de linha a linha no range de 10
vacinas = list()
with open('vacina.csv', encoding = 'iso-8859-1') as arq:
    for i in range(10):
        linha = arq.readline()    
        linha = linha.split(',')
        if linha[25] not in vacinas:
            vacinas.append(linha[25])
    print(vacinas)
```

```python
#Criar lista vazia para armazenar os tipos diferentes de vacinas e percorre toda a lista
vacinas = list()
with open('vacina.csv', encoding = 'iso-8859-1') as arq:
    for linha in arq.readlines():    
        dados = linha.split(',')
        if dados[25] not in vacinas:
            vacinas.append(dados[25])
    print(vacinas)
```

```python
#Contagem de quantas vacinas foram aplicadas no município de São Paulo
contador = 0
with open('vacina.csv', encoding = 'iso-8859-1') as arq:
    for linha in arq.readlines():    
        dados = linha.split(',')
        if dados[9] == 'SAO PAULO':
            contador += 1
    print(contador)
```

```python
#Buscar quais foram os municípios em que se aplicaram vacina
cidades = list()
with open('vacina.csv', encoding = 'iso-8859-1') as arq:
    for linha in arq.readlines():    
        dados = linha.split(',')
        if dados[9] not in cidades:
            cidades.append(dados[9])
    print(sorted(cidades))
```

```python
#Lista de idade de homens e lista de idade mulheres
idade_homem = list()
contador = 0
with open('vacina.csv', encoding = 'iso-8859-1') as arq:
    for linha in arq.readlines():    
        dados = linha.split(',')
        if dados[4] == 'M':
            contador += 1
            idade_homem.append(int(dados[2]))

    print(sum(idade_homem)/contador)
```

```python
#Lista de idade de homens e lista de idade mulheres
idade_mulher = list()
contador = 0
with open('vacina.csv', encoding = 'iso-8859-1') as arq:
    for linha in arq.readlines():    
        dados = linha.split(',')
        if dados[4] == 'F':
            contador += 1
            idade_mulher.append(int(dados[2]))

    print(sum(idade_mulher)/contador)
```
